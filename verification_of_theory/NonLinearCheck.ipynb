```python
import torch.nn as nn
import torch
class NonLinear(torch.nn.Module):
    def __init__(self, x_):
        super().__init__()
        x_ = x_.cpu().numpy()
        xxt = torch.tensor(x_ @ x_.T)
        w = torch.cat([xxt, -xxt], dim=0)
        nin = x.shape[0]
        self.A = nn.Parameter(torch.zeros_like(w.T))
        self.B = nn.Parameter(w)
        self.act = torch.nn.ReLU()
        
    def forward(self, inp):
        return self.A @ self.act(self.B @ inp)
```

```python
import torch
D = 200
x = torch.randn((D, 1)).double()
x = 2*x/torch.linalg.norm(x)
```

```python
net = NonLinear(x).cuda()
x = x.cuda()
```

```python
import torch.optim as optim
optimizer = optim.SGD(net.parameters(), lr=0.1, momentum=0)
train_criterion = torch.nn.MSELoss()
for i in range(10000):
    optimizer.zero_grad()
    inp = torch.cat([x, -x], dim=1) # d x 2
    out = net(inp)
    l = train_criterion(out, inp)
    l.backward()
    optimizer.step()
    if i % 500 == 0:
        print(l.item())
```
